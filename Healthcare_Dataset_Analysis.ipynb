# Data Visualization

## Age Distribution

```python
plt.figure(figsize=(10,6))
sns.histplot(data['Age'], bins=20, kde=True, color='blue')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

## Gender Distribution

```python
plt.figure(figsize=(6,6))
data['Gender'].value_counts().plot(kind='pie', autopct='%1.1f%%', colors=[ 'lightpink','lightblue'])
plt.title('Gender Distribution')
plt.ylabel('')
plt.show()
```

## Blood Type Distribution

```python
blood_type_counts = data['Blood Type'].value_counts()

plt.figure(figsize=(6,6 ))
plt.pie(blood_type_counts, labels=blood_type_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Blood Type Distribution')
plt.axis('equal') 
plt.show()

```

## Admission Type Distribution

```python
admission_type_counts = data['Admission Type'].value_counts()

plt.figure(figsize=(6, 6))
admission_type_counts.plot(kind='bar', color='skyblue')
plt.title('Admission Type Distribution')
plt.xlabel('Admission Type')
plt.ylabel('Count')
plt.xticks(rotation=45) 
plt.show()
```

##  Billing Amount Boxplot

```python
plt.figure(figsize=(8, 6))
sns.boxplot(y='Billing Amount', data=data, color='skyblue')
plt.title('Boxplot of Billing Amount')
plt.ylabel('Billing Amount')
plt.show()
```

##  Average Billing Amount by Admission Type 

```python
avg_billing_by_admission = data.groupby('Admission Type')['Billing Amount'].mean().sort_values()
plt.figure(figsize=(8, 6))
avg_billing_by_admission.plot(kind='bar', color='skyblue')
plt.title('Average Billing Amount by Admission Type')
plt.xlabel('Admission Type')
plt.ylabel('Average Billing Amount')
plt.show()
```

## Most Common Medical Conditions

```python
common_medical_conditions = data['Medical Condition'].value_counts().head(10)
plt.figure(figsize=(8, 6))
common_medical_conditions.plot(kind='bar', color='salmon')
plt.title('Most Common Medical Conditions')
plt.xlabel('Medical Condition')
plt.ylabel('Count')
plt.show()
```

##  Top 5 Hospitals With Highest Billing Amount

```python
top_hospitals = data.groupby('Hospital')['Billing Amount'].sum().sort_values(ascending=False).head(5)
plt.figure(figsize=(8, 6))
top_hospitals.plot(kind='bar', color='lightgreen')
plt.title('Top 5 Hospitals with Highest Billing Amount')
plt.xlabel('Hospital')
plt.ylabel('Total Billing Amount')
plt.show()
```
